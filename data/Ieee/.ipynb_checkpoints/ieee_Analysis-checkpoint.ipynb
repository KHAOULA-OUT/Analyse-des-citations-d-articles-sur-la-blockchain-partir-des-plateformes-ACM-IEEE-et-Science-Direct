```python
import string 
def findCountry(stringText):
    countiers_names = []
    Universities = []
    
    nouns = str(stringText)
    nouns_1 = nouns.split(',')
    print(nouns_1)
    
    for i in nouns_1:
        # delete witespaces from the end and begining from strins
        i = str(i).rstrip()
        i = str(i).lstrip()
        # remove ponctuation
        i = i.translate(str.maketrans('', '', string.punctuation))
        
        # get Universities Names 
        matches = ["University", "university","Universiti","universiti","Universities","universities"]
        if any(x in i for x in matches):
            if not(i in Universities):
                print("condition verified_1______University")
                print(i)
                Universities.append(i)

        # get countries Names
        country = CountryInfo(str(i))
        country_others_names = country.alt_spellings()
        if country_others_names:
            if not(i in countiers_names):
                print("condition verified_1______composed country")
                print(i)
                countiers_names.append(i)
                
    nouns_2 = nouns.split(' ')
    print(nouns_2)
    for i in nouns_2:
        # Remove punctuation from string
        i = i.translate(str.maketrans('', '', string.punctuation))
#         i = i.replace(" ","")
        i = str(i).rstrip()
        i = str(i).lstrip()
        
        country = CountryInfo(str(i))
        country_others_names = country.alt_spellings()
        if country_others_names:
            if not(i in countiers_names):
                print("condition verified_2_____ simple country")
                print(i)
                countiers_names.append(i)
#     print(countiers_names)
    return Universities,countiers_names 
```

```python
def addCountry_Row(row):
    new_rows = []
    Universities,countiers = findCountry(row["Author Affiliations"])
    for country in countiers:
        if country != '':
            df2 = {'Document Title':row["Document Title"],
                   'Authors':row["Authors"],
                   'Author Affiliations':country,
                   'Universities':Universities,
                   'Publication Year':row["Publication Year"],
                   'Abstract':row["Abstract"],
                   'ISSN':row["ISSN"],
                   'ISBNs':row["ISBNs"],
                   'DOI':row["DOI"],
                   'Article Citation Count':row["Article Citation Count"],
                   'Patent Citation Count':row["Patent Citation Count"],
                   'Publisher':row["Publisher"]
                  }
            new_rows.append(df2)
    return new_rows
```

## Citation "Article Citation Count " +	"Patent Citation Count	"

```python
import math

def get_citations(row):
    citation = 0
    article_citation = float(row["Article Citation Count"])
    if math.isnan(article_citation):
        article_citation = 0
    article_citation = float(article_citation)
    print("article_citation_____",article_citation)
    
    
    patent_citation = float(row["Patent Citation Count"])
    if math.isnan(patent_citation):
        patent_citation = 0
    patent_citation = float(patent_citation)
    print("patent_citation_______",patent_citation)
    
    citation = article_citation + patent_citation
    print("citation_______",citation)
    return citation
    
```

```python
echant2 = final_data_2.head()
echant2["citation"] = 0
echant2
```

```python
#Test 
import time
start_time = time.time()

for index, row in echant2.iterrows():
    echant2.loc[index, ['citation']] = get_citations(row)

#     row["citation"] = get_citations(row)
```
