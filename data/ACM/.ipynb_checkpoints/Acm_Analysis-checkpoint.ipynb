```python
import string 
def find_Universities(stringText):
    Universities = []
    
    nouns = str(stringText)
    nouns_1 = nouns.split(',')
    print(nouns_1)
    
    for i in nouns_1:
        # delete witespaces from the end and begining from strins
        i = str(i).rstrip()
        i = str(i).lstrip()
        # remove ponctuation
        i = i.translate(str.maketrans('', '', string.punctuation))
        # get Universities Names 
        matches = ["University", "university","Universiti","universiti","Universities","universities"]

        if any(x in i for x in matches):
            if not(i in Universities):
                print("condition verified_1______University")
                print(i)
                Universities.append(i)
 
    return Universities 
```

## handle date

```python
import datefinder

def get_year(text):
    
    year=0
    try: 
#         date_noised = str(text).split(',')[-2]
#         print("noised_date",date_noised)

        date_noised = str(text).rstrip()
        date_noised = str(text).lstrip()

        date = datefinder.find_dates(str(date_noised))
        for match in date:
            date_string = match
            year_String = str(date_string).split('-')[0]
            year = int(year_String)
            print("year",year)
    except:
        print("no date finded")
        
    return year
```

```python
get_year("19 July 2022")
```

```python
import time
start_time = time.time()

for index, row in data.iterrows():
    data.loc[index, ['Publication Year']] = get_year(row["Publication Year"] )
    
print("--- %s seconds ---" % (time.time() - start_time)) 

#echant
```
