```python
import pandas as pd
```

```python
ScDir_Finale_Data = pd.read_csv("ScDir/ScDir_Finale_Data.csv")
IEEE_Finale_Data = pd.read_csv("Ieee/IEEE_Finale_Data.csv")
Acm_Finale_Data = pd.read_csv("ACM/Acm_Finale_Data.csv")
```

## make the same columns names and order and than join the three tables

```python
ScDir_Finale_Data = ScDir_Finale_Data.loc[:, ~ScDir_Finale_Data.columns.str.contains('^Unnamed')]
ScDir_Finale_Data = ScDir_Finale_Data.iloc[1:,:]
ScDir_Finale_Data.head(1)
```

```python
IEEE_Finale_Data = IEEE_Finale_Data.loc[:, ~IEEE_Finale_Data.columns.str.contains('^Unnamed')]
# IEEE_Finale_Data = IEEE_Finale_Data.iloc[1:,:]
IEEE_Finale_Data = IEEE_Finale_Data.loc[:,['Document Title','Authors','Author Affiliations','Universities','Publication Year','citation','DOI','Publisher']]
IEEE_Finale_Data.columns= ['Document Title','Authors','Author Affiliations','Universities','Publication Year','Citations','DOI','Publisher']
IEEE_Finale_Data.head(1)
```

```python
Acm_Finale_Data = Acm_Finale_Data.loc[:, ~Acm_Finale_Data.columns.str.contains('^Unnamed')]
Acm_Finale_Data = Acm_Finale_Data.loc[:,['Document Title','Authors','Author Affiliations','Universities','Publication Year','citation','DOI','Publisher']]
Acm_Finale_Data.columns= ['Document Title','Authors','Author Affiliations','Universities','Publication Year','Citations','DOI','Publisher']
Acm_Finale_Data.head(1)
```

##  Concatenating the 3 DataFrames 

```python
frames = [ScDir_Finale_Data, IEEE_Finale_Data,Acm_Finale_Data]
  
result = pd.concat(frames)
display(result)
```

```python

def country_flag(df):
    list_alpha_2_1 = [(i.alpha_2).lower() for i in list(pycountry.countries)]
    list_alpha_3_1 = [(i.alpha_3).lower() for i in list(pycountry.countries)]   

    country_code_0 = df['Author Affiliations']
    country_code_0 = str(country_code_0).rstrip()
    country_code_0 = country_code_0.lstrip()
    
    country_code = df['Author Affiliations']
    country_code = str(country_code).lower()
    country_code = country_code.rstrip()
    country_code = country_code.lstrip()
    
    try:
        
        if (country_code == 'vietnam'):
            return "Vietnam"
        
        if (country_code == 'romania'):
            return "Romania"
        
        if (country_code == 'taiwan'):
            return "Taiwan"
        
        if (country_code == 'syria'):
            return "Syria"
        
        if (country_code == 'island'):
            return "Island"
        
        if (country_code == 'russia'):
            return "Russia"
        
        if (country_code == 'nigeria'):
            return "Nigeria"
        
        if (country_code is 'dominican republic'):
            return "Dominican Republic"
        
        if (str(country_code) in ['the netherlands','netherlands']):
            return "Netherlands"
        
        if (str(country_code) in ['Republic of China','P.R. China']):
            return "China"
        
        if (str(country_code) is 'republic of korea','korea','south korea'):
            return "South Korea"
        
        if (str(country_code) is 'united states of america'):
            return "United States"
        
        if (str(country_code) in ['uk','u.k','united kingdoms']):
            return "United Kingdom"
        
        if (len(country_code)==2 and country_code in list_alpha_2_1):
            return pycountry.countries.get(alpha_2=country_code_0).name

        elif (len(country_code)==3 and country_code in list_alpha_3_1):
            return pycountry.countries.get(alpha_3=country_code_0).name
        else:
            return get_country_name(country_code)
        
    except :
        return '0000'
```
